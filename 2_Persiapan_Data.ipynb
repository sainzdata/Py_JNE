# Import Data

```python
import pandas as pd                                             # library untuk impor data
sales = pd.read_excel('data/DP2.xlsx', sheet_name=0)          # simpan sebagai variabel sales
print(sales)                                                  # print data
type(sales)         
```

## Expor Data

```python
sales.to_excel('data/sales.xlsx', sheet_name='bebas2')
```

```python
sales = sales.sort_values("Sales January",ascending=False)  # mengurutkan data (a-z) 
print(sales)                                                # print data      
```

```python
q1_columns = sales[["Sales January", "Sales February", "Sales March"]]
sales["Sales Q1"] = q1_columns.sum(axis=1)
sales
```

```python
managers = pd.read_csv('data/Managers.csv',sep=";")
print(managers)
```

```python
sales = pd.merge(sales, managers, how='left', on='Department')
print(sales)
```

```python
sales["Current Employee"] = pd.isnull(sales['End Date'])
sales.replace(to_replace = [""], value =["2012-10-16"])
```

```python
sales['Department'].value_counts()
```

```python
test= sales.pivot_table(index='Department', 
                  values='Sales Q1', 
                  aggfunc='mean')
test
```
