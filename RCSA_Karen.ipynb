```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

fig = px.histogram(b,x="Departemen", color="STATUS MITIGASI")
fig.show()
```

```python
import plotly.graph_objects as go

colors = ['yellow',] * 5
colors[1] = 'crimson'

fig = go.Figure(data=[go.Bar(
    x=['Not Priority', 'Priority'],
    y=[392, 174],
    textposition='auto',
    marker_color=colors # marker color can be a single color value or an iterable
)])
fig.update_layout(title_text='Status Mitigasi')
```

```python
import plotly.graph_objects as go

colors = ['yellow',] * 5
colors[1] = 'green'
colors[2] = 'blue'
colors[3] = 'red'

fig = go.Figure(data=[go.Bar(
    x=['acceptable', 'Issue', 'Supplementary Issue', 'Unacceptable'],
    y=[198, 127,194, 47],
    textposition='auto',
    marker_color=colors # marker color can be a single color value or an iterable
)])
fig.update_layout(title_text='Status Mitigasi')
```

PIE CHART

```python
label=b['Risk Type'].value_counts().index
count=b['Risk Type'].value_counts().values
color = ["red", "yellow", "green","blue"]

plt.figure(1, figsize=(20,10))
plt.pie(count, labels=label, colors=color, autopct='%1.1f%%')
plt.show()
```
