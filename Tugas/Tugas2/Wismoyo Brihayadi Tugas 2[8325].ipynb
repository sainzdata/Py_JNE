## Latihan Pertemuan 2

```python
import pandas as pd
data_2020 = pd.concat(pd.read_excel('Omset_202001_202012.xlsx', 
                                    sheet_name=None), 
                                    ignore_index=True)
data_2021 = pd.concat(pd.read_excel('Omset_202101_202112.xlsx', 
                                    sheet_name=None), 
                                    ignore_index=True)
data_2022 = pd.concat(pd.read_excel('Omset_202201_202207.xlsx', 
                                    sheet_name=None), 
                                    ignore_index=True)
# gabungkan semua data
all_data = pd.concat([data_2020,data_2021,data_2022],ignore_index=True)
print(all_data)
```
