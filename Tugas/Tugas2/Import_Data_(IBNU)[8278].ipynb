# 2.5.4 8 Baris Pertama (Tugas)
## Perlihatkan 8 baris pertama dari data yang sedang anda gunakan!

```python
data_2020 = pd.concat(pd.read_excel('Omset_202001_202012.xlsx', 
                                    sheet_name=None), 
                                    ignore_index=True)
data_2021 = pd.concat(pd.read_excel('Omset_202101_202112.xlsx', 
                                    sheet_name=None), 
                                    ignore_index=True)
data_2022 = pd.concat(pd.read_excel('Omset_202201_202207.xlsx', 
                                    sheet_name=None), 
                                    ignore_index=True)
# gabungkan semua data
all_data = pd.concat([data_2020,data_2021,data_2022],ignore_index=True)
print(all_data)
```

```python
# menampilkan baris ke 1 sampai ke 8

df= pd.DataFrame (all_data) 

df.loc[0:8]
```

```python
##Lakukan pembulatan untuk setiap nilai desimal yang anda miliki didalam dataset tersebut diatas!

df [['JUMLAH_CASH']] = df [['JUMLAH_CASH']] .astype ('int64')
df [['JUMLAH_CREDIT']] = df [['JUMLAH_CREDIT']] .astype ('int64')
df [['JUMLAH_COD']] = df [['JUMLAH_COD']] .astype ('int64')
df [['JUMLAH_EPAY']] = df [['JUMLAH_EPAY']] .astype ('int64')
df [['TOTAL_OMZET']] = df [['TOTAL_OMZET']] .astype ('int64')
df.tail()       
```

```python
#Pilihlah data berdasarkan beberapa kolom dengan menggunakan string dan number!

df [["REGION", "BRANCH_ID", "BRANCH_NAME"]]
```

```python
#Pilihlah data berdasarkan beberapa kolom dengan menggunakan string dan number!

df [["JUMLAH_AWB", "JUMLAH_CASH", "TOTAL_OMZET"]]
```

```python
# Buatlah kolom baru dengan nilai penjumlahan atau pengurangan atau perkalian dari kolom lainnya.

cash_cod = df [["JUMLAH_CASH", "JUMLAH_COD"]]
df ["Total_CASH&COD"] = cash_cod.sum(axis=1) #nambah kolom Total_Cash&COD baru dan penjumlahan
cr_epay = df [["JUMLAH_CREDIT", "JUMLAH_EPAY"]]
df ["Total_Credit&EPAY"] = cr_epay.sum(axis=1) #nambah kolom Total_Credit&EPAY baru dan penjumlahan
T_C = df [["Total_CASH&COD" , "Total_Credit&EPAY"]]
df ["Total_Cek"] = T_C.sum(axis=1) #nambah kolom Total_Cek baru dan penjumlahan
df
```

```python
#proses sortir data berdarsar atrribut kolom yang anda inginkan!

sortir = df.sort_values("Total_Cek" ,ascending=True)      # mengurutkan data (a-z) 
df [["Total_Cek" , "TOTAL_OMZET"]]                        # print data               
```

```python
# Contoh tabel pivot dengan Python!

df.pivot_table(index='REGION', 
                  values='TOTAL_OMZET', 
                  aggfunc='sum')

df.sort_values("TOTAL_OMZET" ,ascending=True)
```
